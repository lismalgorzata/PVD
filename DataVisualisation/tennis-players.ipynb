The list below contains the tennis players who have won the most Grand Slam tournaments. Create a new notebook and then a chart presenting the data from the table. Remember to include the title and description of the chart axis.

 Player         | Titles
---------------|-------
Novak Djokovic | 24
Rafael Nadal   | 22
Roger Federer  | 20
Pete Sampras   | 14

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titles = [24,22,20,14]
ds = pd.Series(titles)
ds
```

```python
ds.index = ['Novak Djokovic', 'Rafael Nadal', 'Roger Federer', 'Pete Sampras']
ds
```

```python
ds.plot(kind='bar', title="The most Grand Slam winners", xlabel='Player', ylabel='Number of Wins')
```
