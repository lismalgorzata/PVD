average age of women and men (two bars)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
employees = pd.read_csv("employees.csv")
employees
```

```python
employees_selected = employees.loc[:, ['sex', 'age']].groupby(['sex'])
employees_mean = employees_selected['age'].mean()
employees_mean.plot(kind='bar', title='average age of women and men', ylabel='Average age by sex')
```
