#### Wczytaj plik continents.csv do obiektu DataFrame. Następnie wybierz kolumny Continent i Population i utwórz wykres słupkowy przedstawiający populację każdego kontynentu.

```python
import pandas as pd
```

```python
continents = pd.read_csv('continents.csv', index_col=None)
df = pd.DataFrame(continents)
df
```

```python
chart_data = df.loc[:, ['Continent', 'Population'] ]
chart_data.set_index(['Continent'], inplace=True)
chart_data
```

```python
chart_data.plot(kind='barh', title='Population', xlabel= 'People')
```

#### Utwórz obiekt DataFrame na podstawie danych dotyczących kaloryczności posiłków studenta: śniadania (650 kalorii), lunchu (420 kalorii) i kolacji (890 kalorii). Następnie stwórz wykres słupkowy przedstawiający kaloryczność każdego posiłku. Dodaj tytuł wykresu oraz opisy osi X i Y.

```python
meals = pd.DataFrame(
    {
        'meal': ['śniadanie', "lunch", 'kolacja'],
        'calories': [650, 420,890]
    }
)
meals
```

```python
meals.index = ['śniadanie', "lunch", 'kolacja']
meals.plot(kind='bar', title='meal\'s calories', xlabel='meal', ylabel='calories')
```
