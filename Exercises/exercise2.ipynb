1. Stwórz obiekt Series zawierający temperatury w ciągu tygodnia (np. od poniedziałku do niedzieli). Następnie wykonaj następujące czynności:
1. Wybierz temperaturę dla środy.
1. Oblicz średnią temperaturę dla całego tygodnia.
1. Wybierz temperaturę dla dni, w których temperatura przekracza 25 stopni.

```python
temperatures = pd.Series([20, 22, 25, 24, 28, 26, 23], index=['poniedziałek', 'wtorek', 'środa', 'czwartek', 'piątek', 'sobota', 'niedziela'])
temperatures
```

```python
temperature_sroda = temperatures['środa']
print("Temperatura dla środy:", temperature_sroda)
```

```python
srednia_temperatura = temperatures.mean()
print("Średnia temperatura dla całego tygodnia:", srednia_temperatura)
```

```python
temp_przek_25 = temperatures[temperatures > 25]
print("Temperatura dla dni, w których przekracza 25 stopni:")
print(temp_przek_25)
```

1. Stwórz obiekt Series zawierający oceny uczniów z matematyki (np. oceny od 1 do 6). Następnie wykonaj następujące czynności:
1. Oblicz liczbę ocen dostatecznych (ocena co najmniej 3).
1. Oblicz procent uczniów, którzy zaliczyli egzamin (ocena co najmniej 3).
1. Znajdź indeks (numer ucznia) osoby, która uzyskała najwyższą ocenę.
1. Sprawdź, czy jakakolwiek osoba nie zdała egzaminu (ocena 1).

```python
oceny = pd.Series([4, 5, 3, 2, 6, 4, 4, 3, 6, 1], index=['uczeń1', 'uczeń2', 'uczeń3', 'uczeń4', 'uczeń5', 'uczeń6', 'uczeń7', 'uczeń8', 'uczeń9', 'uczeń10'])
oceny
```

```python
liczba_dostatecznych = (oceny >= 3).sum()
print("Liczba ocen dostatecznych:", liczba_dostatecznych)
```

```python
procent_zaliczonych = (oceny >= 3).mean() * 100
print("Procent uczniów, którzy zaliczyli egzamin:", procent_zaliczonych)
```

```python
najwyzsza_ocena_indeks = oceny.idxmax()
print("Indeks osoby, która uzyskała najwyższą ocenę:", najwyzsza_ocena_indeks)
```

```python
czy_zdal_egzamin = (oceny == 1).any()
print("Czy jakakolwiek osoba nie zdała egzaminu (ocena 1):", czy_zdal_egzamin)
```

```python
czy_zdal_egzamin = (oceny == 1).sum()
print(f'{czy_zdal_egzamin} osób nie zdało egzaminu')
```

```python
czy_zdal_egzamin = (oceny != 1).sum()
print(f'{czy_zdal_egzamin} osób zdało egzamin')
```
